Control Spiral 

```python
plt.figure(figsize= (9,9))
for i in range(1, 10, 1):
    plt.subplot(3,3,i)
    img = load_img("../Parkinsons_Data/spiral/training/healthy/"+
                  os.listdir("../Parkinsons_Data/spiral/training/healthy/")[i])
    plt.title("Control Spiral Drawing") 
    plt.imshow(img)   
plt.show() 
```
